```python
df = pd.read_excel('/content/suburbs.xlsx')
```

```python
def clean_string(text):
    text = re.sub('[^A-Za-z0-9\s]', ' ', text)
    return text.title()
```

```python
df['Postcode'] = df['Concatenation for Domain'].apply(lambda x: x[-4:].title())
df['State'] = df['Concatenation for Domain'].apply(lambda x: x.split('-')[-2].title())
df['Suburb Name'] = df['Concatenation for Domain'].apply(lambda x: ' '.join(x.split('-')[:-2]).title())
df.head()
```

```python
df['Suburb Name'] = df['Suburb Name'].apply(lambda x: clean_string(x))
```

```python
df['State'].unique()
```

```python
state_names = {
    'Tas': 'Tasmania',
    'Nsw': 'New South Wales',
    'Qld': 'Queensland',
    'Nt': 'Northern Territory',
    'Vic': 'Victoria',
    'Sa': 'South Australia',
    'Act': 'Australian Capital Territory',
    'Wa': 'Western Australia'
    }
df['State'] = df['State'].replace(state_names)
df.head()
```

```python
df = df.rename(columns={'Concatenation for Domain':'Search string for Domain'})
df['Link for Domain'] = df['Search string for Domain'].apply(lambda x: 'https://www.domain.com.au/suburb-profile/'+str(x))
del df['Concatenation for Homely']
df
```

```python
homely = pd.read_csv('/content/HomelyExtract_20230622.csv')
# df['Postcode'] = df['Concatenation for Domain'].apply(lambda x: x[-4:].title())
# df['State'] = df['Concatenation for Domain'].apply(lambda x: x.split('-')[-2].title())
# df['Suburb Name'] = df['Concatenation for Domain'].apply(lambda x: ' '.join(x.split('-')[:-2]).title())
# df.head()
```

```python
homely.dropna(subset=['Suburb Name'], inplace=True)
```

```python
homely['Postcode'] = homely['Suburb Name'].apply(lambda x: x[-4:].title())
homely['State'] = homely['Suburb Name'].apply(lambda x: x.split(' ')[-2].title())
homely['Suburb'] = homely['Suburb Name'].apply(lambda x: (' '.join(x.split(' ')[:-2]).title()).strip(','))
```

```python
homely['State'] = homely['State'].replace(state_names)
```

```python
homely['State'].unique()
```

```python
homely = homely[homely['State'] != 'Code']
```

```python
homely.drop_duplicates(['Postcode', 'State', 'Suburb'],inplace = True)
homely.shape
```

```python
homely['Number of Reviews'] = homely['Number of Reviews'].str.extract('(\d+)').astype(int)
homely['Search string for Homely'] = homely['Url'].apply(lambda x: x.split('/')[3])
homely['Region'] = homely['Region'].replace(np.nan, 'Na/na/Na/na/na').apply(lambda x: x.split('/')[4])
homely['Region'] = homely['Region'].apply(lambda x: ' '.join(x.split('-')).strip(' ') )

state_list = list(homely.State.unique())
for item in state_list:
    homely['Region'] = homely['Region'].str.replace(item.lower(), '')
homely['Region'] = homely['Region'].str.title()
homely['Region'] = homely['Region'].str.strip(' ')

#homely[homely['Link_check'] == 1]
```

```python
homely['Suburb'] = homely['Suburb'].apply(lambda x: clean_string(x))
```

```python
homely = homely.rename(columns={'Url':'Link for Homely','Suburb Name':'Search String'})
```

```python
homely.columns
```

```python
custom_order = ['Postcode', 'State', 'Suburb', 'Search String', 'Ratings', 'Number of Reviews', 'Rankings',
       'Great for', 'Not great for', 'Who lives here?', 'Time Stamp', 'Region', 'Search string for Homely','Link for Homely']

homely = homely.reindex(columns=custom_order)
```

```python
homely = homely.sort_values(by=['Postcode', 'Suburb'])
homely = homely.reset_index(drop=True)
homely.index += 1
```
